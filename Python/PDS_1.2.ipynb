```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
n = np.arange(-N, N+1) / (N/6.28)

x1 = np.sin(n**2)
plt.plot(n,x1)
plt.title("a) x(n) = sin(n^2) señal par")
plt.show()
```

```python
x2 = np.exp(np.abs(n))
plt.plot(n,x2)
plt.title("a) x(n) = e^|n| señal par")
```

```python
x3 = np.cos((n/2)**2)
plt.plot(n,x3)
plt.title("c) x(n) = cos( (n/2) ^2) Señal par")
plt.show()
```

```python
x4 = n**2 +2
plt.plot(n,x4)
plt.title("d) x(n) = n^2 +2 Funcion par")
plt.show()
```

```python
x5 = np.tanh(n)
plt.plot(n,x5)
plt.title("e) x(n) = tanh(n) Funcion impar")
plt.show()
```

```python
x6 = n**3
plt.plot(n,x6)
plt.title("e) x(n) = n^3 Funcion impar")
plt.show()
```

```python
x7 = n**2 - n
plt.plot(n,x7)
plt.title("e) x(n) = n^2 - n Funcion ni par ni impar")
plt.show()
```
