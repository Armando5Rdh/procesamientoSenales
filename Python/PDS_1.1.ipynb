```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
Omega = 2
Theta = 0
t = np.arange(-2*np.pi,2*np.pi,0.1)
xa = A * np.cos(Omega*t + Theta)

plt.plot(t, xa)
plt.xlabel("t")
plt.ylabel("xa(t)")
plt.title("Señal de tiempo contínuo")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
Omega = 2
Theta = 0
t = np.arange(-2*np.pi,2*np.pi,0.1)
xa = A * np.cos(Omega*t + Theta)
plt.plot(t, xa)
plt.xlabel("t")
plt.ylabel("xa(t)")
plt.title("Señal de tiempo contínuo")

t2 = np.arange(-2*np.pi,2*np.pi,0.1)
xa2 = A * np.exp(1j*(Omega*t2+Theta))
plt.plot(t2, xa2,'o','r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
theta = 0
Omega = 2
t = np.linspace(-2*np.pi, 2*np.pi, 500)
xa = A * np.exp(1j * (Omega*t+theta))

xa_real = xa.real
xa_imag = xa.imag

plt.plot(t,xa_real)
plt.plot(t,xa_imag)
plt.title("Sistema rectangular")
plt.legend(["Parte real","Parte imaginaria"])
plt.xlabel("Tiempo t")
plt.ylabel("Amplitud xa(t)")
plt.show()
```

```python
xa_mag = np.abs(xa)
xa_fase = np.angle(xa)

plt.plot(t,xa_mag)
plt.plot(t,xa_fase)
plt.title("Sistema polar")
plt.legend(["Magnitud","Fase"])
plt.xlabel("Tiempo t")
plt.ylabel("Amplitud xa(t)")
plt.show()
```
